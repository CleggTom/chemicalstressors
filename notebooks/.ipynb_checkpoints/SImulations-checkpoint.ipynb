# modular uptake


```julia
#Varying leakage
N_leak = 10
N_mod = 5
N_rep = 5

N,M = 25, 25

l_vec = range(0.0,0.8,length = N_leak)
mod_vec = range(1.0, 10.0, length = N_mod)

res = Array{Any,5}(undef, N_leak, N_mod, N_mod, N_rep, 2)

for i = 1:N_leak
    for j = 1:N_mod
        for k = 1:N_mod9t
            #define uptake function
            fu(N,M,kw) = MiCRM.Parameters.modular_uptake(N,M, N_clusters = N, s_ratio = mod_vec[j])
            fl(N,M,kw) = MiCRM.Parameters.modular_leakage(M, N_clusters = N, s_ratio = mod_vec[k])

            for r = 1:N_rep
                p = MiCRM.Parameters.generate_params(N, M, f_u = fu , λ = l_vec[i], S_ind = N + M + 1, d = 0.0, γ = rand(N))

                #inital simulation
                x0 = rand(N+M+1)
                x0[p.kw.S_ind] = 0.0
                t = (0.0,1e6)

                prob = ODEProblem(g,x0,t,p)
                res[i,j,k,r,1] =  solve(prob, AutoTsit5(Rosenbrock23()))

                #remove extinct
                prob_ext = MiCRM.Stressors.remove_extinct(res[i,j,k,r,1])
                prob_ext.u0[prob_ext.p.kw.S_ind] = 0.1

                res[i,j,k,r,2] = solve(prob_ext, AutoTsit5(Rosenbrock23()))
                end
        end
    end
end
```

```julia
u1 = res[1,1,1,1,1].prob.p.u
u2 = res[1,end,1,1,1].prob.p.u

d1 = pairwise(cosine_dist, u1,u1, dims=1)
d2 = pairwise(cosine_dist, u2,u2, dims=1)

f = Figure(resolution = (400,400))
ax1 = Axis(f[1,1])
ax2 = Axis(f[1,2])
ax3 = Axis(f[2,1])
ax4 = Axis(f[2,2])

heatmap!(ax1, u1)
heatmap!(ax2, d1)

heatmap!(ax3, u2)
heatmap!(ax4, d2)

f
```

# interactions

```julia
function mean_offdiag(A)
    A_tot = sum(A) - sum(diag(A))
    return(A_tot / (dim(A) * (dim(A) - 1)))
    
end

#calculate overalap
function avg_comp_overlap(sol)
    X = sol.prob.p.u
    R = 1 .- pairwise(cosine_dist, X, X, dims=1)
    
    return(mean_offdiag(R))
end

#calculate overalap
function avg_coop_overlap(sol)
    X = sol.prob.p.u
    Y = sol.prob.p.u * sol.prob.p.l
    R = 1 .- pairwise(cosine_dist, X, Y, dims=1)
    
    return(mean_offdiag(R))
end
```

# System responses

```julia
# System responses
function δN(sol1, sol2)
    N1 = sum(sol1[end] .> eps()) - sol1.prob.p.M - 1
    N2 = sum(sol2[end] .> eps()) - sol2.prob.p.M - 1
    return(N2 - N1)
end

function δC(sol1, sol2)
    C1 = sum(sol1[end][1:sol1.prob.p.N]) 
    C2 = sum(sol2[end][1:sol2.prob.p.N])
    return(C2 - C1)
end

function δR(sol1, sol2)
    R1 = sum(sol1[end][sol1.prob.p.N + 1 : end-1]) 
    R2 = sum(sol2[end][sol2.prob.p.N + 1 : end-1])
    return(R2 - R1)
end
```

```julia
#Apply metrics
comp = map(avg_comp_overlap, res[:,:,:,:,1])
coop = map(avg_coop_overlap, res[:,:,:,:,1])

ΔN = [δN(res[i,j,k,r,1],res[i,j,k,r,2]) for i = 1:N_leak, j = 1:N_mod, k = 1:N_mod, r = 1:N_rep]
ΔC = [δC(res[i,j,k,r,1],res[i,j,k,r,2]) for i = 1:N_leak, j = 1:N_mod, k = 1:N_mod, r = 1:N_rep]
ΔR = [δR(res[i,j,k,r,1],res[i,j,k,r,2]) for i = 1:N_leak, j = 1:N_mod, k = 1:N_mod, r = 1:N_rep];
```

## multigaussian kernel

```julia
N = 10
M = 20

θC = rand(N)
# sort!(θC)
θM = range(0,1,length  = M)

u = abs.(θC .- θM')
u = mapslices(x -> x/sum(x) , u, dims = 2)

dist = pairwise(cosine_dist, u, u, dims=1)
```

```julia
function quad_trap(f,a,b,N) 
    h = (b-a)/N
    int = h * ( f(a) + f(b) ) / 2
    for k=1:N-1
        xk = (b-a) * k/N + a
        int = int + h*f(xk)
    end
    return int
end

function gaussian(x,b,a)
    exp(-((x - b)^2) / a^2)
end

#sum mulitple functions
function f_sum(funcs, x)
    res = 1.0
    for f = funcs
        res -= f(x)
    end
    return(max(0,res))
end

o = -1.0

function sample_multiguass(μ,σ; N = 1, o = -1.0)
    #initial function
    funcs = [x -> gaussian(x, i,j) for (i,j) in zip(μ,σ)]

    #get summed function
    f_summed(x) = f_sum(funcs, x)

    norm = quad_trap(f_summed,0,1,100) 

    f_final(x) = f_summed(x) / norm
    
    return(sample(θ,  Weights(f_final.(0:0.001:1.0)), N))
end
```

```julia
## Modular construction
N = 50
M = 50

N_clusters = 10

s_ratio = 10.0

function modular_uptake(N,M,kw)
    N_clusters = kw[:Nclust]
    s_ratio = kw[:sratio]
    
    @assert N_clusters <= M
    
    cluster_size = M ÷ N_clusters
    u = zeros(N,M)
    
    #assign guilds equal
    for g = 1:N_clusters
        indx = 1 + (g-1)*cluster_size : cluster_size + (g-1)*cluster_size

        if g == N_clusters
            indx = 1 + (g-1)*cluster_size : M
        end
        #calculate prob vec
        a_vec = ones(M)
        a_vec[indx] .*= s_ratio

        d = Dirichlet(a_vec)
        # println(size())
        x = rand(d, length(indx))

        u[indx, :] .= x'
    end
    
    return(u)
end
```

```julia
#uptake functions
N_clusters = 5
s_ratio = 10.0

function modular_leakage(M; N_clusters = 2, s_ratio = 10.0, λ = 0.5)
    @assert N_clusters <= M
    
    #cluster  sizes for consumers and resources
    cluster_R = M ÷ N_clusters

    #preallocate leakage matrix
    l = ones(M,M)

    #loop over clusters
    for g = 1:N_clusters
        for h = 1:N_clusters
            indx_R1 = 1 + (g-1)*cluster_R : (g == N_clusters ? M : cluster_R + (g-1)*cluster_R)
            indx_R2 = 1 + (h-1)*cluster_R : (h == N_clusters ? M : cluster_R + (h-1)*cluster_R)

            if h == g+1
                #calculate prob vec
                l[indx_R1, indx_R2] .*= s_ratio
            end
        
        end
    end

    for i = 1:M
        l[i,:] = rand(Dirichlet(l[i,:]), 1)
    end

    return(l .* λ)
end

l = modular_leakage(100,N_clusters = 10, s_ratio = 10)

l |> heatmap
```
