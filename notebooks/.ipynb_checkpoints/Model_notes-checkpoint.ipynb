# Introduction

In this notebook I set out the plan for simulating the effects of stressors on microbial communties. 
    
I start with the basic formulation of the MiCRM describing the dynamics of $N$ bacterial consumer populations and $M$ resources:
\begin{align}
    \frac{dC_i}{dt} &=  \sum_{\alpha}^M \bigg[ C_i R_{\alpha} u_{i \alpha} (1 - \sum_{\beta}^{M} l_{\alpha \beta}) \bigg] - C_i m_i, \\
    \frac{dR_{\alpha}}{dt} &= \rho_{\alpha} - R_{\alpha} \omega_{\alpha} - \sum_{i}^N C_i R_{\alpha} u_{i \alpha} + \sum_{i}^N \sum_{\beta}^M C_i R_{\beta} u_{i \beta} l_{\beta \alpha}.
\end{align}



The question is how to include the effects of stressors. I will do this by including a stressor varible in the system of equations $S$. We then need to define 1) how the stressor dynamics evolve over time and 2) how the stressor affects the consumer populations. 

### Stressor Dynamics
In simplest case stressors are independent of the consumers with a given inflow-outflow rate:
$$
\frac{dS}{dt} = \rho_S - S \omega_S.
$$
This gives a press type disturbance with a constant equiblirium concentration in the environment $\frac{\rho_S}{\omega_S}$.

We can also consider a pulse disturbance where the the supply function is discontinous in time:
\begin{align}
\frac{dS}{dt} = f(t) - S \omega_S \ \ \text{where} \ \ f(t) = \begin{cases} 0 \ \text{if} \ t \notin t_{stops} \\
                                                                            s \ \text{if} \ t \in t_{stops} 
                                                              \end{cases}
\end{align}

where the stressor is supplied in a given quantitiy at times $t_{stops}$.

### Consumer responses
Here I will consider only the simple case where stressors impose a direct mortality on consumers. This is equivelent to antibiotics whose mode of action is to break down cell walls. One could include responses such as uptake inhibition or reductions in biomass synthesis (in the leakage term?). The direct mortality term $D_i$ has a very simple form:
$$
    D_i(S) = d_i S C_i
$$

where the reduction in mass-specific growth is given by the concentration of the stressor mulitplied by a sensitvtiy of that strain to the stressor. 

### Where do the dead cells go?
The final peice of this is to decide where the resources tied up in cell biomass go with this extra mortality? There are two options here: 1) these resources leave the system (as with the existing maintenece/mortality $m_iC_i$) and 2) cell mass goes back into the system with a (fixed?) distribution of resources. 

The second of these would require a new term in the resource dynamic equation A:
$$
A_{\alpha} = \sum_{i}^N D_i (S) a_{\alpha}
$$

where the vector $a$ is constrained as $\sum a = 1$ and represents the resource composition of the cellular mass in terms of avalible resources. 

## A new model
Overall the new equations become:

\begin{align}
    \frac{dC_i}{dt} &=  \sum_{\alpha}^M \bigg[ C_i R_{\alpha} u_{i \alpha} (1 - l_{\alpha}) \bigg] - C_i m_i - C_i d_i S, \\
    \frac{dR_{\alpha}}{dt} &= \rho_{\alpha} - 
                                R_{\alpha} \omega_{\alpha} - 
                                \sum_{i}^N C_i R_{\alpha} u_{i \alpha} +
                                \sum_{i}^N C_i d_i S a_{\alpha} + 
                                \sum_{i}^N \sum_{\beta}^M C_i R_{\beta} u_{i \beta} l_{\beta \alpha} \\
    \frac{dS}{dt} &= f(t) - \omega_S S
\end{align}


# Multiple Stressors

We can also conside the effects of multiple chemical stressors and their interactive effects. To do so we need only consider additional stressors (i.e. $S_1, S_2, \ldots, S_O$) which have the same dynamic equations:
$$
\frac{dS_\alpha}{dt} = f_{\alpha}(t) - \omega^{S}_{\alpha}
$$


To consider interactions we can consider disturbance as a function of these multiple stressors and encode the desired interactions:

$$
    D(\vec{S}) = C_i(d_1 S_1 + d_2 S_2 + d_{1:2} S_1 S_2),
$$

# Parameterisation

In general the model setup above gives us 3 main parameters of interest
1) Uptake matrix - how consumers differ in their uptakes
2) Leakage matrix - How the production of byprodcuts overlaps
3) Sensitvtiy vector - How consumers vary in their sensitvtiy to the stressors. 

The remaining parts we can hold constant?
* resource dynamics (inflow-outflow)
* stressor dynamics (inflow-outflow)
* consumer mortality

```julia
using Pkg
Pkg.activate("..")

using MiCRM
using Plots
```

```julia
#lazy way to add stressors. 
N = 5
M = 10

p = MiCRM.random_micrm_params(N,M,0.4)
@named sys = MiCRM.micrm_system(p)

#get eqns and statevars ect. 
eqns = equations(sys)
params = parameters(sys)
state = states(sys)

#add stressor to system
@variables t S(t) 
@parameters ρₛ ωₛ
d = ones(N)

D = Differential(t)
s_eqn = D(S)  ~ ρₛ - ωₛ * S

#add to original system
[eqns[i] = eqns[i].lhs ~ eqns[i].rhs - state[i] * S * d[i] for i = 1:N]

push!(eqns, s_eqn)
@named stress_sys = ODESystem(eqns, t);
```

```julia
function stressor_sim(x)
    #generate parameter set
    params = ModelingToolkit.default_p(sys)
    params[ρₛ] = x
    params[ωₛ] = 1.0
    #Starting biomass set

    #convert to problem
    #define starting mass
    u0 = ones(N+M+1)
    u0 = [states(stress_sys)[i] => u0[i] for i = eachindex(u0)]

    tspan = (0.0, 1e6) #define tspan
    prob = ODEProblem(stress_sys,u0,tspan,params, jac = true)
    sol = solve(prob)
end
```

```julia
S_vec = 0.1:0.2:2.0
@time x = stressor_sim.(S_vec);
```

```julia
end_mat = hcat([y[end] for y = x]...)'

plot(log.(max.(0,end_mat[:,1:N])))
```

```julia
#square with sign
sq_sgn(x) = sign(x) * x^2

function euclid_dist(a,b, sign = false)
    sign ? sum(sq_sgn, a .- b) : sum(abs2, a .- b)
end

function u_func(d)
    x = exp(-100d)
    
    x < 0.1 ? 0.0 : x
end

D = 1
N = 100
M = 100

# function niche_micrm_params(N,M,leakage, D)
    #uniform random niches
    θC = [rand(D) for i = 1:N]
    θR = [rand(D) for i = 1:M]

    indx_C = sortperm(norm.(θC,1))
    indx_R = sortperm(norm.(θR,1))

    #calculate consumer resource distance matrix
    d_mat_u = [euclid_dist(x,y) for x = θC, y = θR][indx_C,indx_R]
    #caluclate leakage distance matrix
    d_mat_l = [euclid_dist(x,y, true) <= 0 ? Inf : euclid_dist(x,y)  for x = θR, y = θR][indx_R,indx_R]

    #convert to actual matricies
    u_mat = u_func.(d_mat_u)
    l_mat = u_func.(d_mat_l)

heatmap(l_mat)
```
